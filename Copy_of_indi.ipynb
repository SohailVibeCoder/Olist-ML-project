```python
#monthly sharpe ratio
crsp_factors['excess'] = crsp_factors['ret'] - crsp_factors['rf']

sharpe_stats = crsp_factors.groupby("permno")["excess"].agg(["mean", "std"]).reset_index()

sharpe_stats.columns = ["permno", "mean_excess", "std_excess"]

sharpe_stats["monthly_sharpe"] = sharpe_stats["mean_excess"] / sharpe_stats["std_excess"]


print("Monthly Statistics:")
print(sharpe_stats.head())
```

```python
#annual Sharpe Ratio
sharpe_stats["annual_sharpe"] = sharpe_stats["monthly_sharpe"] * np.sqrt(12)

Table_B4 = sharpe_stats[["permno", "annual_sharpe"]].copy()

Table_B4.columns = ["permno", "annualised_sharpe_ratio"]

print(Table_B4)
```

```python
crsp_factors.head()
```

```python
crsp_factors['mkt_cap'] = crsp_factors['prc'] * crsp_factors['shrout']


crsp_factors = crsp_factors.sort_values(['permno', 'yearmonth'])

# Lagged Market Cap (t-1)

crsp_factors['lag_mkt_cap'] = crsp_factors.groupby('permno')['mkt_cap'].shift(1)

# Drop rows where lagged market cap is missing (e.g., the very first month of the sample)

vw_stats = crsp_factors.dropna(subset=['lag_mkt_cap']).copy()

# 5. Calculate Total Market Cap for each Month (the denominator)

monthly_total_cap = vw_stats.groupby('yearmonth')['lag_mkt_cap'].transform('sum')

# 6. Calculate Dynamic Monthly Weights
vw_stats['weight'] = vw_stats['lag_mkt_cap'] / monthly_total_cap

# 7. Calculate Weighted Excess Returns
# Excess return for the stock * Weight of the stock
vw_stats['excess'] = vw_stats['ret'] - vw_stats['rf']
vw_stats['vw_data_excess'] = vw_stats['weight'] * vw_stats['excess']

# 8. Aggregate to Portfolio Level
# Summing the weighted returns of all stocks for a specific month gives the portfolio return
pmonthly = vw_stats.groupby("yearmonth")['vw_data_excess'].sum()

# 9. Calculate Performance Metrics
avg_monthly = pmonthly.mean()
vw_sharpe = (avg_monthly / pmonthly.std()) * np.sqrt(12)

print(avg_monthly, " average excess returns of the weighted portfolio")
print(vw_sharpe, " sharpe ratio of weighted portfolio")
```
